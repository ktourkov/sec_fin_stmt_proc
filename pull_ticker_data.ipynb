```python
import pandas as pd
import pandas_datareader.data as web
import datetime as dt
```

```python
start = dt.datetime(2015, 1, 1)
end = dt.datetime.now()
```

```python
df = web.DataReader("AAPL", 'yahoo', start, end)
```

```python
df1 = df.reset_index()
df1['qtr_end'] = df1['Date'] + pd.tseries.offsets.QuarterEnd(n=0)
df1 = df1.sort_values(by='Date')
df1 = df1.drop_duplicates(['qtr_end'],keep='last')\
         .drop(['Date','High','Low','Open','Volume','Adj Close',], axis=1)
display(df1)
```
